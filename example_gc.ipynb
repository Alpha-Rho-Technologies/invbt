**Import Backtester Code:**

```python
from inv_bt.backtester import bt, pd, dt
```

**Import Backtester Files:**

```python
from os.path import join
from os import getcwd

# Set current working directory path as base directory:
FILE_PATH = join(getcwd(),'investment_strategy_backtester','files')

# load Asset price data parameters:
PORTFOLIOS = pd.read_csv(join(FILE_PATH,'portfolios.csv'),index_col=0)

# Transform columns to datetime format:
PORTFOLIOS.columns = pd.to_datetime(PORTFOLIOS.columns)

# Load portfolios at different points in time:
ASSET_PRICE_DATA = pd.read_csv(join(FILE_PATH,'apd.csv'),index_col='Date',parse_dates=True)
```

**Example Backtest: S&P 500 and Dow Jones 30 Portfolio rebalanced every 2 Months:**

```python
ENDING_DATE = dt.date(2023,8,31)
TRANSACTION_COSTS = 0.00015
BALANCE_FREQ = 'M'
STARTING_BALANCE = 100

backtest = bt(portfolios=PORTFOLIOS,
              apd=ASSET_PRICE_DATA,
              balance_freq=BALANCE_FREQ,
              end_date=ENDING_DATE,
              trans_cost=TRANSACTION_COSTS,
              starting_balance=STARTING_BALANCE
              )
```

Transform Backtest Output to Dataframe:

```python
bt_df = pd.concat(backtest.values())
```

Visualize Performance:

```python
bt_df.plot(title= 'Backtest Balance')
```
